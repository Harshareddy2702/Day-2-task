## Dictionary

```python
# retrieving the key from the dictionary
dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
print(dict1['CSK'])
print(dict1['MI'])
```

```python
# Assigning new key and value to the existing dictionary

dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
dict1['SRH']='WARNER'
print(dict1)
```

```python
# Updating new value to the existing key
dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
dict1['RCB']="ABD"
print(dict1)
```

```python
# Deleting the key and value in the dictionary

dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
del dict1['RCB']
print(dict1)
```

```python
#Dictionary allows all types of data types
dict1={
    "num1": "[2,10]",
    "set1":"{hello}",
    "tup1":"(10,20)"
}
print(dict1)
```

```python
#dictionary does not allow duplicate keys amd it is immutable
# in dictionaries tuple and keys are immutable
dict1={
    "num1":"[2,10]",
    "set1":"{hello}",
    "tup1":"(10,20)",
    "num1":"[2,10]"
}
print(dict1)
```

```python
# Clear() in dictionary
dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
dict1.clear()
print("dict1 is:",dict1)
```

```python
#copy() in dictionary
dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
new=dict1.copy()
print("original dict is :",dict1)
print("new dict is :",new)
```

```python
# get() function in dictionary
dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
print("RCB capitan is : ",dict1.get("RCB"))
```

```python
#items() function in dictionary
dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
print("original dict :",dict1)
del[dict1["RCB"]]
print("new dict",dict1.items())
```

```python
#keys() function in dictionary
dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
key=dict1.keys()
print(key)
```

```python
# pop() function in dictionary
dict1={
    "CSK" : "M S DHONI",
    "KXIP" : "K L RAHUL",
    "MI" : "ROHIT",
    "RR" : "SAMSON",
    "RCB" : "V KOHLI"
}
dict1.pop("RCB")
print(dict1)
```

## Sets

```python
s={"DHONI","K L RAHUL","ROHIT","KOHLI"}
print(s)
```

```python
# add function in sets
s={"DHONI","K L RAHUL","ROHIT","KOHLI"}
s.add("BHUVI")
s.add("JADEJA")
print(s)
```

```python
# update function in sets
s={"DHONI","K L RAHUL","ROHIT","KOHLI"}
t={"JADEJA","RAINA"}
s.update(t)

print(s)
```

```python
# remove function in sets
s={"DHONI","K L RAHUL","ROHIT","KOHLI"}
s.remove("KOHLI")
print(s)
```

```python
#discard function in sets
s={"DHONI","K L RAHUL","ROHIT","KOHLI"}
s.discard("KOHLI")
print(s)
```
