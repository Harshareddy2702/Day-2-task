```python
import PyPDF2

temp = PyPDF2.PdfFileReader(open(r'G:\harsha reddy\harsha pdfs\unit 1 COA theory.pdf','rb'))
water = PyPDF2.PdfFileReader(open(r'G:\harsha reddy\harsha pdfs\unit 3 COA theory.pdf','rb'))

output = PyPDF2.PdfFileWriter()

for i in range(temp.getNumPages()):
    page = temp.getPage(i)
    page.mergePage(water.getPage(0))
    output.addPage(page)

    with open(r'G:\harsha reddy\harsha pdfs\unit 3 COA theory.pdf','wb') as file:
        output.write(file)
```
