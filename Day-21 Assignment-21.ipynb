## Map(func,iter)

```python
def func(n) :
    return n+10
numbers=(1,2,3,4)
results=map(func,numbers)
print(list(results))
```

```python
def func(a,b) :
    return a+" "+b

a=["Harsha","Nikhil","Kamalakar"]
b=["reddy","reddy","reddy"]
results=map(func,a,b)
print(tuple(results))
```

## Filter()

```python
def func(variables):
    letters=['a','e','i','o','u']
    if (variables in letters) :
        return True
    else :
        return False
# sequence 
sequence = ['g', 'e', 'e', 'j', 'k', 's', 'p', 'r'] 

# using filter function 
filtered = filter(func, sequence) 
print('The filtered letters are:') 
for s in filtered: 
    print(s) 

    
```

```python
ages=[20,1,5,18,35,40]
def func(x):
    if x<=18:
        return False
    else:
        return True
adults= filter(func, ages)
for x in adults:
    print(x)
```

## Zip()

```python
def func(a,b):
    return a+b
a=['Air','Blow','Catch']
b=['Port','Out','Up']
c=zip(a,b)
print(list(c))
```

## Reduce()

```python
import functools
lis=[1,20,5,6,]
print("The Sum of the list is :",end= " ")
print(functools.reduce(lambda a, b:a+b,lis))

print("The Maximum length of the list is :",end=" ")
print(functools.reduce(lambda a, b:a if a>b else b,lis))
```
