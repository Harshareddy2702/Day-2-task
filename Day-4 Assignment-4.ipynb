## Numeric Type - int,float,complex

```python
a=15.5
b=20
c=10
d=a+b+c

print(int(d))
print(float(d))
print(complex(d))
```

## Sequence Type - List,Tuple,Range

```python
my_list=[1,2,4,5,["Hello"]]
print(my_list)
print(my_list[4])
print(len(my_list))
print("Let's continue the Tuple in next line")
```

```python
# Tuple - It can be any Data Type
my_tuple=("hi","hello",[1,2],1,2.5) # Strings must be in quotations
print(my_tuple)
```

```python
# range() - It prints the Start from start value till stop
x=range(1, 100, 2)
for n in x :
    print(n)
```

## Mapping Type - dictionary

```python
my_dict={
    "Fruit":"Mango",
    "Vegetable":"Tomato",
    "Flower":"Rose"
}
print(my_dict)
print(my_dict["Fruit"])
print(len(my_dict))
print(type(my_dict))
```

## Set Types - Sets,Forzenset

```python
# Set Types - Sets,Frozen Sets
my_set={"hi","hello",(3,2),"hi"} # In Sets Duplicates will not Allowed
print(my_set)
```

```python
# Frozenset
nu=(1,2,3,4,5,"hi")
fnum=frozenset(nu)
print("The Forzenset is :" ,fnum)
```
